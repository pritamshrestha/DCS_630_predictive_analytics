```python
# diplaying each day of week confirmed cases
# creating function to sum up the each day value
def add(date):
    day=[]
    for i in date:
        day.append(i)
        r=sum(day)
    return r
    
day1_c=add(df_us_c['6/16/2020'])  
day2_c=add(df_us_c['6/17/2020'])  
day3_c=add(df_us_c['6/18/2020'])  
day4_c=add(df_us_c['6/19/2020'])  
day5_c=add(df_us_c['6/20/2020'])  
day6_c=add(df_us_c['6/21/2020'])  
day7_c=add(df_us_c['6/22/2020'])  
# Printing all values
print(day1_c)
print(day2_c)
print(day3_c)
print(day4_c)
print(day5_c)
print(day6_c)
print(day7_c)
```
